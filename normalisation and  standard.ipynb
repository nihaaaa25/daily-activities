```python
import pandas as pd
s=pd.Series([1000,2000,3000,5000,9000])
s
```

# normalisation

```python
s.min()
```

```python
s.max()
```

```python
snorm=(s-s.min())/(s.max()-s.min()) #formula
print(snorm) 
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
import pandas as pd
s=pd.Series([1000,2000,3000,5000,9000])
s
```

```python
snorm=MinMaxScaler().fit_transform(s.values.reshape(-1,1)) # Using 
print(snorm)
```

```python
df=pd.DataFrame({
    "temp":[20,50,34,56],
    "Humidity":[56,21,32,34]
})
df
```

```python
df.min()
```

```python
df.max()
```

```python
norm_data=MinMaxScaler().fit_transform(df) # how to normalise data
norm_data
```

```python
norm_df=pd.DataFrame(norm_data) #converting array into dataframe
norm_df
```

```python
norm_df=pd.DataFrame(norm_data,columns=df.columns) #it gives a column names.
norm_df
```
