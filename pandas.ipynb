# concate Data Frame

```python
india_weather=pd.DataFrame({
    "city":["mumbai","banglore","hyderabad"],
    "temperature":[34,23,44]
})
india_weather
```

```python
us_weather= pd.DataFrame({
    "city":["chicago","Newyork","orland"],
    "temperature":[45,12,76]
})
us_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True) # it gives line by line ignore index true means it follows ordered index
df
```

```python
df=pd.concat([india_weather,us_weather],axis=1) # it gives side by side tabless
df 
```

# Merge dataframes

```python
temp_df=pd.DataFrame({
    "city":["hybd","vzg","vijaywada"],
    "temperature":[34,56,21]
})
temp_df
```

```python
hum_df=pd.DataFrame({
    "city":["chicago","Newyork","vzg"],
    "humidity":[45,12,76]
})
hum_df
```

```python
df=pd.merge(temp_df,hum_df,on="city") # it merges and gives common clm
df 
```

```python
df=pd.merge(temp_df,hum_df,on="city",how="outer") # it gives all with null values
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["temperature"]=df["temperature"].fillna(df.temperature.min())
```
