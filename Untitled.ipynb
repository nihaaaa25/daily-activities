#  DATA PIPELINE
          A data pipeline is a step by step process that takes raw data and converts it into model ready data.
### Basic flow:
          Raw data---> Cleaniing--->processing--->model
### Why do we need a pipeline
          Real world / Raw data data is not clean:
              > Missing values
              > Duplicate records
              > Different formats
              > Noise and outliers
### A data pipeline:
          > Oraganises data properly
          > Automates steps
          > Reduce human errors
          > Makes ML modelm reliable
### Pipeline in sklearn:
         A pipeline in sklearn is a sequence of data preproocessing steps and a machine learning model combined together 
         to automate training and prediction.
### simple data pipeline flow:
         Data collection--->Data preperation--->ML model training--->ML model evaluation--->ML model deployement
### Common sources:
         CSV / Excel files
         APIs
         ex: student marks data collected from a csv file
### Data Cleaning:
         Fixing errors and problems in data.
### Common task:
         Handling missing values
         Removing duplicates
         Fixing incorrect values
         removing outliers
         
         ex: Fill missing study hours with average
             Removve duplicates student records
### Data preprocessing:
         Converting data into a machine understandable format.
### Common techniques:
         > scaling ( StandardScaler, MinMaxScaler)
         > Encoding ( Label Encoding, One-hot Encoding)
         > Train Test Split
         
         ex: Convert "Male/Female"-->0/1
             Scale marks to same range.
### Feature engineering:
         Creating new and better features from existing data.
### Common methods:
         > Binning 
         > Feature creation
         > Feature selection
         > Feature transformation
         Ex: Convert age--> Young/Adult/Senior
             Create "total marks" from subjects marks
### Model Training:
         Giving processed data to a machine learning  algorithms so it can learn patterns.
### Common algorithms:
         Linear regression
         Logistic regressio
         decision tree
         random forest
### Model evaluation:
         Checking how well the model is performing.
              > accuracy
              > Mean squared error(MSE)
              > mean absolute error(MAE)
              > R^2 Score
              
              EX: compare predicted marks with actual marks.
